```python
# filtering rows using conditions in pandas
import pandas as pd
data={
    "Id":[101,102,103,104,105],
    "Name":["alice","charlie","eva","david","bob"],
    "Age":[22,25,23,24,26],
    "marks":[85,78,92,88,90]
}
df=pd.DataFrame(data)
print("original dataframe:")
print(df)
#filter marks greater than 85
print("\n students with marks>85:")
print(df[df["marks"]>85])
#filter age less than 25
print("\n students with age <25:")
print(df[df["Age"]<25])
```

```python
#data manipulation operations
import pandas as pd
data={
    "Name":["A","B","c"],
    "Age":[20,21,22],
    "marks":[70,80,90]

}
df=pd.DataFrame(data)
print("original data:")
print(df)
#add a new column
df["bonus"]=5
print("\n after adding bonus:")
print(df)
#increasing marks
df["marks"]=df["marks"]+2
print("\n after increasing marks:")
print(df)
#create new column from existing column
df["total"]=df["marks"]+df["bonus"]
print(df)
#change value
df.loc[0, "Age"]=21
print("\n after changing age of first student:")
print(df)
#rename column
df.rename (columns={"total":"final"},inplace=True)
print("after renaming total to final")
print(df)
# remove a column
df.drop("bonus",axis=1,inplace=True)
print(df)
```

```python
# grouping of data
import pandas as pd
data = {
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT'],
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [20, 21, 19, 22, 18],
    'Marks': [85, 78, 92, 88, 90]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
group_mean = df.groupby('Department')['Marks'].mean()
print("\nAverage Marks by Department:")
print(group_mean)
group_stats = df.groupby('Department')['Marks'].agg(['mean', 'max', 'min'])
print("\nMarks Statistics by Department:")
print(group_stats)
group_count = df.groupby('Department')['Name'].count()
print("\nStudent Count by Department:")
print(group_count)
group_multi = df.groupby(['Department', 'Age'])['Marks'].mean()
print("\nGrouped by Department and Age:")
print(group_multi)
```

```python
# Selection Sort Program
arr = [45, 25, 117, 22, 11]
n = len(arr)
for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
print("Sorted array:")
print(arr)
```

```python
# Bubble Sort Program
arr = [77, 34, 25, 44, 22, 65, 90]
n = len(arr)
for i in range(n - 1):
    for j in range(n - i - 1):
        if arr[j] > arr[j + 1]:
            # swap
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
print("Sorted array:")
print(arr)
```

```python
# Linear Search Program
arr = [15, 25, 30, 44, 66]
key = 30
found = False
for i in range(len(arr)):
    if arr[i] == key:
        print("Element found at index:", i)
        found = True
        break
if not found:
    print("Element not found")
```

```python
# Binary Search Program
arr = [10, 20, 30, 40, 50, 60]
key = 40
low = 0
high = len(arr) - 1
found = False
while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print("Element found at index:", mid)
        found = True
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1
if not found:
    print("Element not found")
```

```python
#rearranging data
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 28],
    'Salary': [50000, 60000, 45000, 52000]
}
df = pd.DataFrame(data)
df_sorted = df.sort_values(by='Age')
print("Sorted by Age:\n", df_sorted)
df_rearranged = df[['Salary', 'Name', 'Age']]
print("\nColumns rearranged:\n", df_rearranged)
df_multi_sorted = df.sort_values(by=['Age', 'Salary'], ascending=[False, True])
print("\nSorted by Age desc, Salary asc:\n", df_multi_sorted)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 92, 85, 78]
}
df = pd.DataFrame(data)
df['Rank'] = df['Score'].rank(ascending=False, method='min')
df_sorted = df.sort_values(by='Rank')
plt.figure(figsize=(8,5))
plt.bar(df_sorted['Name'], df_sorted['Score'], color='skyblue')
plt.title('Scores of Students')
plt.xlabel('Name')
plt.ylabel('Score')
plt.show()
```
