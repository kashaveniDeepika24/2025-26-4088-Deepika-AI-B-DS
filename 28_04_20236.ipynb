```python
from sklearn.linear_model import Lasso
import numpy as np
x=np.array([[1,25,160],[2,30,165],[3,35,170]])
y=np.array([30000,40000,50000])
model=Lasso(alpha=0.1)
model.fit(x,y)
print(model.coef_)
```

```python
from scipy.spatial import distance
from scipy.stats import pearsonr
import numpy as np
p1 = np.array([5,7])
p2 = np.array([2,3])
euc = distance.euclidean(p1,p2)
eh = distance.cityblock(p1,p2)
ch = distance.chebyshev(p1,p2)
hm = distance.cosine(p1,p2)
cs = 1-ch
set1,set2 = set(p1),set(p2)
js = len(set1&set2)|len(set1&set2)
r=pearsonr(p1,p2)
print("Euclidean distance",euc)
print("Cityblock",eh)
print("Manhattan distance",hm)
print("Chebyshev distance",ch)
print("cosine distance",cs)
print("Jascord distance",js)
print("password coeffecient",r)
```

```python
import numpy as np
A = np.array([[2,3],[3,4],[4,2]])
B = np.array([[8,7],[9,6],[10,8]])
P = np.array([5,6])
CA = A.mean(axis=0)
CB = B.mean(axis=0)
dA = np.linalg.norm(P-CA)
dB = np.linalg.norm(P-CB)
if dA<dB:
    print("points belong to cluster A")
else:
    print("points belong to cluster B")

```
