```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
#dataset
x=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([30,35,50,55,65])
#create a model
model=LinearRegression()
#train model
model.fit(x,y)
#predict values
predications=model.predict(x)
#print slope and intercept
print("slope:",model.coef_)
print("intercept:",model.intercept_)
#plot graph
plt.scatter(x,y)
plt.plot(x,predications,color="red")
plt.xlabel("experience")
plt.ylabel("salary")
plt.title("Linear Regression")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

url = "https://raw.githubusercontent.com/ybiFoundation/Dataset/main/Salary%20Data.csv"

# load dataset
data = pd.read_csv(url)
print("Dataset preview")
print(data.head())

x = data[['Experience Years']]
y = data['Salary']

model = LinearRegression()
model.fit(x, y)

print("\n model parameters")
print("slope:", model.coef_[0])
print("intercept:", model.intercept_)

years = [[5]]
predicted_salary = model.predict(years)

print("\n predicted salary for 5 years experience:")
print(predicted_salary[0])

y_pred = model.predict(x)
mse = mean_squared_error(y, y_pred)

print("\n mean squared error:", mse)

plt.plot(x, y_pred)
plt.scatter(x, y)
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.title("Linear Regression salary prediction")
plt.show()
```

```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score

# Sample dataset
data = pd.DataFrame({
    'message': [
        "Win money now", "Hello friend", "Free offer!!!",
        "Let's meet tomorrow", "Claim your prize", "How are you?"
    ],
    'label': [1, 0, 1, 0, 1, 0]  # 1 = Spam, 0 = Not Spam
})

# Convert text to numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data['message'])
y = data['label']

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

print("Spam Detection Accuracy:", accuracy_score(y_test, y_pred))

# Test new message
msg = ["Free money offer"]
msg_vec = vectorizer.transform(msg)
print("Prediction (1=Spam, 0=Not Spam):", model.predict(msg_vec)[0])
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Sample dataset
data = pd.DataFrame({
    'age': [25, 45, 35, 50, 23, 40],
    'bp': [120, 140, 130, 150, 110, 135],
    'disease': [0, 1, 0, 1, 0, 1]  # 1 = Disease
})

X = data[['age', 'bp']]
y = data['disease']

# Model
model = LogisticRegression()
model.fit(X, y)

# Predict for new patient
new_patient = [[30, 125]]
prediction = model.predict(new_patient)

print("Disease Prediction (1=Yes, 0=No):", prediction[0])
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

# Sample dataset
data = pd.DataFrame({
    'review': [
        "I love this product", "This is amazing",
        "Worst experience ever", "I hate it",
        "Very good", "Not good"
    ],
    'sentiment': [1, 1, 0, 0, 1, 0]  # 1 = Positive
})

# Convert text
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data['review'])
y = data['sentiment']

# Train
model = LogisticRegression()
model.fit(X, y)

# Test
review = ["I really love it"]
review_vec = vectorizer.transform(review)

print("Sentiment (1=Positive, 0=Negative):", model.predict(review_vec)[0])
```
